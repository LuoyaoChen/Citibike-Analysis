```python
import geopandas
import geoplot
import pandas as pd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python
stations = pd.read_csv('intersection_station_cluster.csv')
boroughs = geopandas.read_file(geoplot.datasets.get_path('nyc_boroughs'))
```

```python
crs={'init':'epsg:4326'}
geo_df=geopandas.GeoDataFrame(stations,crs=crs,geometry=geopandas.points_from_xy(stations["longitude"], stations["latitude"]))
```

```python
fig,ax=plt.subplots(figsize=(15,15))
boroughs.plot(ax=ax,alpha=0.4,color="grey")
geo_df[geo_df['cluster']== 1].plot(ax=ax,markersize=100, alpha=0.6,color="green", label="stations type 1")
geo_df[geo_df['cluster']== 2].plot(ax=ax,markersize=100, alpha=0.6,color="darkorange", label="stations type 2")
geo_df[geo_df['cluster']== 3].plot(ax=ax,markersize=100, alpha=0.6,color="yellow", label="stations type 3")
geo_df[geo_df['cluster']== 4].plot(ax=ax,markersize=100, alpha=0.6,color="blue", label="stations type 4")
geo_df[geo_df['cluster']== 5].plot(ax=ax,markersize=100, alpha=0.7,color="skyblue", label="stations type 5")
geo_df[geo_df['cluster']== 0].plot(ax=ax,markersize=100, alpha=0.6,color="purple", label="stations type 6")
plt.xlim((-74.05, -73.8))
plt.ylim((40.63,40.87))
plt.legend()
```
